### Importing packages

```python
import os

import pandas as pd
```

### Read the CSV file

```python
FILE_NAME = "SFS KPI_08-2024.csv"
PATH_INPUT_FILE = f"inputs/{FILE_NAME}"
```

```python
df = pd.read_csv(PATH_INPUT_FILE)
df.head()
```

### Delete rows

```python
df.drop(df[df["[IsGrandTotalRowTotal]"] == True].index, inplace=True)  # noqa: E712
df.head()
```

### Drop columns

```python
drop_columns = [
    "[IsGrandTotalRowTotal]"
]
```

```python
df.drop(columns=drop_columns, inplace=True)
df.head()
```

### Renaming columns

```python
renames_dict = {
    "Geography[Zone]":"[Zone]",
    "Geography[Sub Zone]":"[SubZone]",
    "Geography[Affiliates]":"[Affiliates]",
}
```

```python
df.rename(columns = renames_dict, inplace = True)
df.head()
```

### Insert columns

```python
new_columns = [
    "[NberNoShopActivity]",
    "[PerNoShopActivity]",
    "[NberNotDeclaredShopActivity]",
    "[PerNotDeclaredShopActivity]",

    "[NberNoFoodServicesActivity]",
    "[PerNoFoodServicesActivity]",
    "[NberNotDeclaredFoodServicesActivity]",
    "[PerNotDeclaredFoodServicesActivity]",

    "[NberNoHotBeverageActivity]",
    "[PerNoHotBeverageActivity]",
    "[NberNotDeclaredHotBeverageActivity]",
    "[PerNotDeclaredHotBeverageActivity]",

    "[NberNoCarServicesActivity]",
    "[PerNoCarServicesActivity]",
    "[NberNotDeclaredCarServicesActivity]",
    "[PerNotDeclaredCarServicesActivity]",

    "[NberNoCarWashActivity]",
    "[PerNoCarWashActivity]",
    "[NberNotDeclaredCarwashActivity]",
    "[PerNotDeclaredCarwashActivity]",

    "[NberNoServicesActivity]",
    "[PerNoServicesActivity]",
    "[NberNotDeclaredServicesActivity]",
    "[PerNotDeclaredServicesActivity]"
]
```

```python
for column in new_columns:
    df.insert(loc=1, column=column, value=None, allow_duplicates=False)
df.head()
```

### Reorganized columns

```python
reordered_columns = [
    "[Zone]",
    "[SubZone]",
    "[Affiliates]",
    "[TotalSites]",
    "[TotalSitesActive]",
    "[NberShopActivity]",
    "[PerShopActivity]",
    "[NberNoShopActivity]",
    "[PerNoShopActivity]",
    "[NberNotDeclaredShopActivity]",
    "[PerNotDeclaredShopActivity]",
    "[NberFoodServicesActivity]",
    "[PerFoodServicesActivity]",
    "[NberNoFoodServicesActivity]",
    "[PerNoFoodServicesActivity]",
    "[NberNotDeclaredFoodServicesActivity]",
    "[PerNotDeclaredFoodServicesActivity]",
    "[NberHotBeverageActivity]",
    "[PerHotBeverageActivity]",
    "[NberNoHotBeverageActivity]",
    "[PerNoHotBeverageActivity]",
    "[NberNotDeclaredHotBeverageActivity]",
    "[PerNotDeclaredHotBeverageActivity]",
    "[NberCarServicesActivity]",
    "[PerCarServicesActivity]",
    "[NberNoCarServicesActivity]",
    "[PerNoCarServicesActivity]",
    "[NberNotDeclaredCarServicesActivity]",
    "[PerNotDeclaredCarServicesActivity]",
    "[NberCarWashActivity]",
    "[PerCarWashActivity]",
    "[NberNoCarWashActivity]",
    "[PerNoCarWashActivity]",
    "[NberNotDeclaredCarwashActivity]",
    "[PerNotDeclaredCarwashActivity]",
    "[NberServicesActivity]",
    "[PerServicesActivity]",
    "[NberNoServicesActivity]",
    "[PerNoServicesActivity]",
    "[NberNotDeclaredServicesActivity]",
    "[PerNotDeclaredServicesActivity]"
]
```

```python
df = df.reindex(columns=reordered_columns)
print(df.columns)
df.head()
```

### Export to CSV with same input file name

```python
OUTPUT_DIR = "outputs"
PATH_OUTPUT_FILE = f"{OUTPUT_DIR}/{FILE_NAME}"
```

```python
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
```

```python
df.to_csv(PATH_OUTPUT_FILE, index=False)
```
